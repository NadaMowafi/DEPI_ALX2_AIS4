## Data Visualization:

### 1) read titanic data

```python
import pandas as pd

titanic = pd.read_csv(r'titanic')
titanic.head()
```

### 2) Draw the bar plot:

```python
import matplotlib.pyplot as plt
import seaborn as sns   

plt.figure(figsize=(8,5))
sns.countplot(x='Sex', hue='Survived', data=titanic)
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

### 3) Histogram

```python
plt.figure(figsize=(8,5))
sns.histplot(titanic['Age'], bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

### 4) pie:

```python
# Count of each class
class_counts = titanic['Pclass'].value_counts()
labels = ['Class 1', 'Class 2', 'Class 3']

plt.figure(figsize=(6,6))
plt.pie(class_counts, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#66b3ff','#99ff99','#ff9999'])
plt.title('Distribution of Passenger Classes')
plt.show()
```

### 5) Box Plot:

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Pclass', y='Fare', data=titanic)
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

### 6) scatter plot

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=titanic)
plt.title('Fare vs Age by Survival')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```
